Preprosesing

data dibagi 2, data test dan lebel hasil
data tes diambil dari kolom Pclass, Sex,Sibsp dan Parch

```
# DataTitanic=DataTitanic[['Pclass','Sex','SibSp','Parch']]
X=np.asarray(DataTitanic[['Pclass','Sex','SibSp','Parch']])
# DataTitanic
sex= preprocessing.LabelEncoder()
sex.fit(['female','male'])
X[:,1]=sex.transform(X[:,1]) 
# X
Y=np.asarray(DataTitanic['Survived'])
# Y
```

normalisasi

```
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

pembuatan data dengan komposisi 9/1 data set dan data train

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=0.1, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

pembuatan model dengan logistik reg

```
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

penampilan hasil

```
yhat = LR.predict(X_test)
yhat
```

persentase untuk setiap kasusnya

```
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

hasil persision dan recall

```
from sklearn.metrics import classification_report, confusion_matrix

print (classification_report(y_test, yhat))
```
